```python
X = df.drop('Customer_Segment', axis=1) #features
y = df['Customer_Segment'] #target
```

```python
#Standardize the features 
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

### Explained variance ration - 
- Tells us how much of the total variance in the dataset is captured by adding each successive principal component (PC) in PCA

#### pca.explained_variance_ratio_ : array where each entry represents the amount of variance captured by each principal component 
#### cummulative sum of the variance ratio helps us decide how many PC's to keep 

```python
cummulative_variance_ratio = np.cumsum(pca.explained_variance_ratio_)
```

```python
#plotting the cummulative ratio
plt.figure(figsize=(10, 6))
sns.lineplot(x=range(len(cummulative_variance_ratio)), y=cummulative_variance_ratio)
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance Ratio')
plt.title('Cumulative Explained Variance Ratio vs. Number of Components')
plt.grid(True)
plt.show()
```

```python
print(f"Cummulative variance: \n {cummulative_variance_ratio}")
```

```python
#select number of components that explain 95% of the variance
n_component = np.argmax(cummulative_variance_ratio >= 0.95) + 1
print(f"No of components: {n_component}")
```

```python
pca = PCA(n_components=n_component)
X_pca = pca.fit_transform(X_scaled)
```

```python
#create a new dataframe wiht the pca results 
pca_df = pd.DataFrame(data = X_pca, columns = [f'PC{i+1}' for i in range(n_component)])
pca_df['Customer_Segment'] = y
```

```python
pca_df.head()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=pca_df, x='PC1', y='PC2', hue='Customer_Segment', palette='viridis', s=60)
plt.title('PCA Result (2D)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Customer Segment')
plt.show()
```

```python
pca = PCA(n_components=3)
small_pca = pca.fit_transform(X_scaled)
small_df = pd.DataFrame(data=small_pca, columns=['PC1', 'PC2', 'PC3'])
small_df['Customer_Segment'] = df['Customer_Segment']

sns.pairplot(data=small_df, hue='Customer_Segment')
plt.suptitle('Scatter plots of first three principal components')
plt.show()
```
